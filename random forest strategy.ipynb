```python


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from ucimlrepo import fetch_ucirepo

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import (accuracy_score, classification_report,
                              confusion_matrix, ConfusionMatrixDisplay)


# STEP 1: DATASET LOAD KARO

print(" Dataset load ho raha hai UCI se...")
dry_bean = fetch_ucirepo(id=602)

X = dry_bean.data.features   # 16 features
y = dry_bean.data.targets    # Class column

# y ek DataFrame hai, Series mein convert karo
y = y.values.ravel()

print(f"✅ Dataset load ho gaya!")
print(f"   Features shape : {X.shape}")   # (13611, 16)
print(f"   Target shape   : {y.shape}")   # (13611,)
print(f"\n🫘 Bean Classes: {np.unique(y)}")

# ================================================================
# STEP 2: EXPLORATORY DATA ANALYSIS (EDA)
# ================================================================
print("\n" + "="*50)
print("📊 EDA — Dataset ka Overview")
print("="*50)

# Basic info
print(X.describe().round(2))

# Class distribution — kitni beans har type mein hain?
class_counts = pd.Series(y).value_counts()
print("\n🔢 Class Distribution:")
print(class_counts)

# Visualization — class distribution
plt.figure(figsize=(10, 4))
sns.barplot(x=class_counts.index, y=class_counts.values, palette='Set2')
plt.title('Dry Bean — Class Distribution (Kitni beans har type mein?)')
plt.xlabel('Bean Type')
plt.ylabel('Count')
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig('class_distribution.png', dpi=150)
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(X.corr(), annot=False, cmap='coolwarm', center=0,
            linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=150)
plt.show()

# ================================================================
# STEP 3: PREPROCESSING
# ================================================================
print("\n" + "="*50)
print("⚙️  Preprocessing")
print("="*50)

# Label Encoding — bean names ko numbers mein convert karo
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print(f"Class mapping:")
for i, cls in enumerate(le.classes_):
    print(f"  {cls} → {i}")

# Missing values check
print(f"\nMissing values: {X.isnull().sum().sum()}")  # 0 honi chahiye

# Features already numerical hain, scaling optional hai RF ke liye
# Lekin dikhate hain (good practice)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# ================================================================
# STEP 4: TRAIN-TEST SPLIT
# ================================================================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded,
    test_size=0.2,        # 80% train, 20% test
    random_state=42,
    stratify=y_encoded    # Har class ka proportion same rakho
)

print(f"\n✅ Train size : {X_train.shape[0]} samples")
print(f"✅ Test size  : {X_test.shape[0]} samples")

# ================================================================
# STEP 5: RANDOM FOREST MODEL — FULL PARAMETERS
# ================================================================
print("\n" + "="*50)
print("🌲 Random Forest Training...")
print("="*50)

rf = RandomForestClassifier(

    # ── Ensemble Parameters ───────────────────────────────
    n_estimators=200,         # 200 trees banao
                              # ↑ zyada = better accuracy (but slow)
                              # Try range: 100–500

    # ── Tree Structure Parameters ─────────────────────────
    max_depth=20,             # Har tree max 20 levels deep jaye
                              # None = fully grown (overfitting risk)
                              # Kam = simpler, generalized model

    min_samples_split=5,      # Node ko split karne ke liye min 5 samples chahiye
                              # ↑ zyada = simpler tree

    min_samples_leaf=2,       # Leaf node mein min 2 samples hone chahiye
                              # ↑ zyada = smoother boundaries

    max_leaf_nodes=None,      # Max leaf nodes per tree
                              # None = unlimited

    # ── Feature Sampling (Randomness) ────────────────────
    max_features='sqrt',      # Har split pe sqrt(16) ≈ 4 features consider
                              # 'sqrt' → classification ke liye best
                              # 'log2' → alternative
                              # 1.0   → sab features (bagging only)

    # ── Bagging Parameters ───────────────────────────────
    bootstrap=True,           # Bootstrap sampling use karo (Bagging ON)
                              # False → sab data har tree ko milega

    max_samples=None,         # Bootstrap sample size
                              # None = full dataset size
                              # 0.8  = 80% data per tree

    oob_score=True,           # Out-of-Bag score (free validation!)
                              # Jo samples kisi tree ne use nahi kiye,
                              # unpe test karo

    # ── Class Imbalance ──────────────────────────────────
    class_weight='balanced',  # Dataset thoda imbalanced hai
                              # Balanced: rare classes ko zyada weight

    # ── Performance ──────────────────────────────────────
    n_jobs=-1,                # Saare CPU cores use karo
    random_state=42,          # Reproducible results
    verbose=1                 # Training progress dikhao
)

# Model Train Karo
rf.fit(X_train, y_train)
print(f"\n✅ Training complete!")
print(f"🎯 OOB Score (free validation): {rf.oob_score_ * 100:.2f}%")

# ================================================================
# STEP 6: PREDICTION & EVALUATION
# ================================================================
print("\n" + "="*50)
print("📊 Model Evaluation")
print("="*50)

y_pred = rf.predict(X_test)
y_prob = rf.predict_proba(X_test)  # Probabilities

# Overall Accuracy
acc = accuracy_score(y_test, y_pred)
print(f"\n🏆 Test Accuracy: {acc * 100:.2f}%")

# Detailed Classification Report
print("\n📋 Classification Report (har class ke liye):")
print(classification_report(
    y_test, y_pred,
    target_names=le.classes_
))

# ================================================================
# STEP 7: CONFUSION MATRIX
# ================================================================
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10, 8))
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=le.classes_
)
disp.plot(cmap='Blues', colorbar=True)
plt.title('Confusion Matrix — Dry Bean Random Forest', fontsize=14)
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('confusion_matrix_drybean.png', dpi=150)
plt.show()

# ================================================================
# STEP 8: FEATURE IMPORTANCE
# ================================================================
print("\n" + "="*50)
print("🏆 Feature Importance")
print("="*50)

importances = rf.feature_importances_
feat_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
}).sort_values('Importance', ascending=False)

print(feat_df.to_string(index=False))

plt.figure(figsize=(10, 6))
colors = sns.color_palette('viridis', len(feat_df))
sns.barplot(data=feat_df, x='Importance', y='Feature', palette='viridis')
plt.title('Feature Importance — Kaunsa feature sabse important hai?')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.savefig('feature_importance_drybean.png', dpi=150)
plt.show()

# ================================================================
# STEP 9: CROSS VALIDATION
# ================================================================
print("\n" + "="*50)
print("🔄 5-Fold Cross Validation")
print("="*50)

# Faster model for CV
rf_cv = RandomForestClassifier(
    n_estimators=100, max_depth=20,
    n_jobs=-1, random_state=42
)

cv_scores = cross_val_score(
    rf_cv, X_scaled, y_encoded,
    cv=5,           # 5 folds
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

print(f"\nCV Scores (5 folds): {[f'{s*100:.2f}%' for s in cv_scores]}")
print(f"Mean CV Accuracy   : {cv_scores.mean()*100:.2f}%")
print(f"Std Deviation      : {cv_scores.std()*100:.2f}%")

# ================================================================
# STEP 10: HYPERPARAMETER TUNING (GridSearchCV)
# ================================================================
print("\n" + "="*50)
print("🔧 Hyperparameter Tuning (GridSearchCV)")
print("="*50)

# Chhota grid (fast tuning ke liye)
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'max_features': ['sqrt', 'log2']
}

grid_search = GridSearchCV(
    estimator=RandomForestClassifier(
        n_jobs=-1, random_state=42, oob_score=False
    ),
    param_grid=param_grid,
    cv=3,               # 3-fold CV (fast)
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)

print("GridSearch chal raha hai... (thoda waqt lagega)")
grid_search.fit(X_train, y_train)

print(f"\n🏆 Best Parameters: {grid_search.best_params_}")
print(f"🎯 Best CV Score  : {grid_search.best_score_*100:.2f}%")

# Best model se prediction
best_rf = grid_search.best_estimator_
best_pred = best_rf.predict(X_test)
best_acc = accuracy_score(y_test, best_pred)
print(f"✅ Best Model Test Accuracy: {best_acc*100:.2f}%")

# ================================================================
# STEP 11: EK NAYA BEAN PREDICT KARO
# ================================================================
print("\n" + "="*50)
print("🫘 Naye Bean ka Prediction")
print("="*50)

# Ek sample bean ka data (first test sample)
sample = X_test.iloc[[0]]
true_label = le.inverse_transform([y_test[0]])[0]
predicted_label = le.inverse_transform(rf.predict(sample))[0]
probabilities = rf.predict_proba(sample)[0]

print(f"Actual Bean Type   : {true_label}")
print(f"Predicted Bean Type: {predicted_label}")
print(f"\nProbabilities har class ke liye:")
for cls, prob in zip(le.classes_, probabilities):
    bar = '█' * int(prob * 30)
    print(f"  {cls:12s}: {prob*100:5.1f}% {bar}")
```
